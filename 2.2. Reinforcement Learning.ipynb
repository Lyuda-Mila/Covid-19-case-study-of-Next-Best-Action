## Reinforcement Learning

```python
# import packages
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
np.random.seed(seed=28)
import random
from random import random, randint
```

```python
# read dataset
df = pd.read_csv('Data/data_2020_2021_after_preprocessing.csv', parse_dates=[5])
print(df.shape)
print(df.columns)
df.head()
```

```python
#Since the goal is to descrease growth in cases/deaths, variables were multiplied by -1
df['ConfirmedCasesPctChangeWeekly16_'] = df['ConfirmedCasesPctChangeWeekly16']*(-1)
df['ConfirmedDeathsPctChangeWeekly30_'] = df['ConfirmedDeathsPctChangeWeekly30']*(-1)
# remain only required columns
df_rl = df[['AreaName','Date' ,
            'C1_School closing',
            'C2_Workplace closing',
            'C3_Cancel public events',
            'C4_Restrictions on gatherings',
            'C5_Close public transport', 
            'C6_Stay at home requirements',
            'C7_Restrictions on internal movement',
            'C8_International travel controls', 
            'H1_Public information campaigns', 
            'ConfirmedCasesPctChangeWeekly16_', 'ConfirmedDeathsPctChangeWeekly30_']]


df_rl = df_rl.reset_index()
df_rl = df_rl.drop(columns = ['index'],axis = 1)
print(df_rl.shape)
df_rl.head()
```

```python
# Define state as combination of values C1-C8, H1
c1 = df_rl['C1_School closing']
c2 = df_rl['C2_Workplace closing']
c3 = df_rl['C3_Cancel public events']
c4 = df_rl['C4_Restrictions on gatherings']
c5 = df_rl['C5_Close public transport']
c6 = df_rl['C6_Stay at home requirements']
c7 = df_rl['C7_Restrictions on internal movement']
c8 = df_rl['C8_International travel controls']
h1 = df_rl['H1_Public information campaigns']
state = []
for i in range(0, df_rl.shape[0]):
    st = str(c1[i])[0] + str(c2[i])[0] + str(c3[i])[0] + str(c4[i])[0] + str(c5[i])[0] + str(c6[i])[0] + str(c7[i])[0] + str(c8[i])[0] + str(h1[i])[0]
    state.append(st)

print(len(state)) 
df_rl['state'] = state
df_rl['next_state'] = df_rl.groupby('AreaName')['state'].shift(-1)

state_space = df_rl['state'].unique()
state_space.sort()
print(len(state_space))

df_rl['next_ConfirmedCasesPctChangeWeekly16_'] = df_rl.groupby('AreaName')['ConfirmedCasesPctChangeWeekly16_'].shift(-1)
df_rl['next_ConfirmedDeathsPctChangeWeekly30_'] = df_rl.groupby('AreaName')['ConfirmedDeathsPctChangeWeekly30_'].shift(-1)
df_rl = df_rl.dropna(subset=['next_ConfirmedCasesPctChangeWeekly16_']) # Drop all which do not have a value fo the largest shift
df_rl = df_rl.dropna(subset=['next_ConfirmedDeathsPctChangeWeekly30_']) # Drop all which do not have a value fo the largest shift
df_rl = df_rl.reset_index()
df_rl = df_rl.drop(columns = ['index'],axis = 1)
df_rl = df_rl.sort_values(by = ['Date', 'AreaName'])
```

## RL for ConfirmedCasesPctChangeWeekly16

```python
#function to get q value
def get_q_value(alpha, discount_factor, state_current, action_selected, qtable, real_reward ):
    max_reward_in_new_state = qtable[qtable.index==action_selected].max().max()
    q_current = qtable.loc[state_current, action_selected]
    q_updated = (1-alpha)*q_current + alpha * (real_reward + discount_factor * max_reward_in_new_state)
    return q_updated
```

```python
%%time
# Iterate over qtable based on historical data
state = df_rl['state']
ConfirmedCasesPctChangeWeekly16_ = df_rl['ConfirmedCasesPctChangeWeekly16_']
next_state = df_rl['next_state']
next_ConfirmedCasesPctChangeWeekly16_ = df_rl['next_ConfirmedCasesPctChangeWeekly16_']
#alpha = 0.9
#discount_factor = 0.1
reward_expected = []
reward_real = []
episodes = []
alphas = []
discount_factors = []
for alpha in [ 0.9]:
    for discount_factor in [ 0.1]:
        Z = np.zeros((675,675))#only states we have in dataset. Total number of states = 129600, the matrix will be too heavy
        qtable_cases = pd.DataFrame(Z)
        colnames = dict(list(enumerate(state_space)))
        qtable_cases = qtable_cases.rename(columns=colnames, index=colnames)
        for i in range(0, df_rl.shape[0]):
                episodes.append(i+1)
                q_updated = get_q_value(alpha, discount_factor, state[i], next_state[i], qtable_cases, next_ConfirmedCasesPctChangeWeekly16_[i] )
                reward_real.append(next_ConfirmedCasesPctChangeWeekly16_[i])
                reward_expected.append(q_updated)
                alphas.append(alpha)
                discount_factors.append(discount_factor)
                qtable_cases.loc[state[i], next_state[i]] = q_updated
```
